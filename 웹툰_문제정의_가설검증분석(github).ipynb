```python
import pandas as pd


df = pd.read_csv('네이버웹툰_데이터_최종.csv', index_col=0) 
df['작가 관심수_변동'] = round(df['작가 관심수_변동'])
 
df['월_연재요일'][df['작가명']=='김대영']=1
 

check_yj = df[['작가명','월_연재요일', '화_연재요일', '수_연재요일',  '목_연재요일', '금_연재요일', '토_연재요일', '일_연재요일']].drop_duplicates()
# 연재요일 컬럼 생성
weekdays = ['월', '화', '수', '목', '금', '토', '일']
check_yj['연재요일'] = check_yj.apply(lambda row: [day for day, value in zip(weekdays, row[1:8]) if value == 1], axis=1)
df= pd.merge(df, check_yj[['작가명','연재요일']], left_on ='작가명', right_on ='작가명')
df['연재요일'] = df['연재요일'].astype('str')

df['연재횟수'] = 0
df['연재횟수'][df['연재요일']=="[]"] = '연재x'
df['연재횟수'][(df['연재요일']=="['수', '토']") |(df['연재요일']=="['수', '일']") | (df['연재요일']== "['목', '토']") |(df['연재요일']==  "['화', '금']") |(df['연재요일']== "['월', '수']") |(df['연재요일']== "['목', '일']") ] = '주2회'
df['연재횟수'][df['연재횟수']==0] ='주1회'


# 컬럼 추가 --'게시글 내용_톤(말투)', '게시글 내용_감정',  '게시글 내용_질문 여부' /// 전자 두개는 chat gpt 활용
## 컬럼 추가 : 게시글 내용_글자수 
df = df.reset_index(drop=True)
df['게시글 내용']= df['게시글 내용'].str.replace('더보기','')
df['게시글 내용_글자수'] = df['게시글 내용'].apply(len)
df['게시글 내용_글자수'].describe()
```

```python
#### 월별 증가량 어쩌구 이건 보류 하자 
 

import pandas as pd
import numpy as np
import plotly.graph_objects as go 

m_imsi = df[['월','작가명','작가 관심수_변동']]

author_i_df = pd.DataFrame()

for author_name in m_imsi['작가명'].unique():
    M1 = m_imsi[m_imsi['작가명']==author_name] 
    M1.drop(['작가명'],axis=1 , inplace= True)
 
    M1['월'] =M1['월'].dt.to_timestamp()
    # 월별로 데이터 인덱스 생성
    M1.set_index('월', inplace=True)

    # 월별로 결측값 채우기 (interpolate 사용)
    M1 = M1.resample('M').mean()  # 월별로 리샘플링하여 결측값을 포함한 월별 데이터로 만듬
        # 2024-10까지 월을 추가
    last_date = pd.Timestamp('2024-10-31')
    if M1.index[-1] < last_date:
        # 필요한 날짜 범위 추가 (2024-10까지)
        date_range = pd.date_range(M1.index[-1] + pd.Timedelta(days=1), last_date, freq='M')
        additional_dates = pd.DataFrame(index=date_range, columns=M1.columns)
        M1 = pd.concat([M1, additional_dates])
  
    M1['작가 관심수_변동'] = M1['작가 관심수_변동'].interpolate(method='linear')  # 선형 보간법을 사용하여 결측값 채우기

    M1['작가명'] = author_name
    author_i_df = pd.concat([author_i_df, M1])

author_i_df = author_i_df.reset_index()
author_i_df.columns = ['월', '작가 관심수_변동','작가명']
author_i_df['월'] = pd.to_datetime(author_i_df['월']).dt.to_period('M')

# 작가별 월별 관심수 변동 증가량 계산
author_i_df['관심수_변동_증가량'] = author_i_df.groupby('작가명')['작가 관심수_변동'].diff()

# 결과 출력
author_i_df= author_i_df[author_i_df['관심수_변동_증가량'].notnull()]


# 월별로 집계 (작가 관심수 변동 합계)
df_monthly = author_i_df.groupby('월').agg({'관심수_변동_증가량': 'sum'}).reset_index()
df_monthly['게시글 날짜'] = df_monthly['월'].dt.to_timestamp()  # 월 period를 timestamp로 변환

# 추세선 계산 (1차 회귀)
x_monthly = (df_monthly['게시글 날짜'] - df_monthly['게시글 날짜'].min()).dt.days  # 날짜를 숫자로 변환
y_monthly = df_monthly['관심수_변동_증가량']
coefficients_monthly = np.polyfit(x_monthly, y_monthly, 1)  # 1차 회귀분석
trendline_monthly = np.polyval(coefficients_monthly, x_monthly)  # 추세선 계산

# Plotly 그래프 생성
fig = go.Figure()

# 월별 데이터 추가
fig.add_trace(go.Scatter(
    x=df_monthly['게시글 날짜'],
    y=df_monthly['관심수_변동_증가량'],
    mode='lines+markers',
    name="월별 관심수 변동 (누계)",
    marker=dict(size=8)
))

# 추세선 추가 (붉은색)
fig.add_trace(go.Scatter(
    x=df_monthly['게시글 날짜'],
    y=trendline_monthly,
    mode='lines',
    name='추세선',
    line=dict(dash='solid', color='red', width=3)  # 실선 스타일, 붉은색
))

# 그래프 레이아웃 설정
fig.update_layout(
    title="월별 관심수_변동_증가량 및 추세선 (증가속도)",
    xaxis_title="월",
    yaxis_title="관심수",
    template="plotly_white",
    title_x=0.5,
    legend_title="범례"
)

# 기울기 출력
slope_monthly = coefficients_monthly[0]
print(f"월별 추세선의 기울기: {slope_monthly}")

# 그래프 표시
fig.show() 

```

```python
# 1) 댓글수에 대한 고정효과 모델

# 고정효과 모델 (작가별 고유 특성을 통제) 
model_comments = PanelOLS(
    y_comments,  # 종속변수
    X,           # 독립변수 (설명변수)
    time_effects=True,
    entity_effects= False,
    drop_absorbed=True,
    check_rank=False
)
result_comments = model_comments.fit()
 
result_comments 
```

```python
import pandas as pd
 
results_df = pd.DataFrame({
    'Coefficient': result_comments.params,
    'Std Error': result_comments.std_errors,
    'P-value': result_comments.pvalues,
    'T-stat': result_comments.tstats
})
 
significant_results = results_df[results_df['P-value'] < 0.05]
significant_results
```

```python
import pandas as pd

results_df = pd.DataFrame({
    'Coefficient': result_reactions.params,
    'Std Error': result_reactions.std_errors,
    'P-value': result_reactions.pvalues,
    'T-stat': result_reactions.tstats
})

# 결과 출력 
significant_results = results_df[results_df['P-value'] < 0.05]
significant_results
```

```python
import pandas as pd
from scipy import stats

# 대조군  / 실험군 :이미지 있는거  (우리가 알고자 하는거) 

experimental_group_comment = df[df['이미지 여부'] ==1]['댓글수'] # 실험군
control_group_comment = df[df['이미지 여부'] ==0]['댓글수'] # 대조군
print(len(experimental_group_comment), len(control_group_comment))
print('')

experimental_group_reaction = df[df['이미지 여부'] ==1]['반응수'] # 실험군
control_group_reaction = df[df['이미지 여부'] ==0]['반응수'] # 대조군
print(len(experimental_group_reaction), len(control_group_reaction))
print('')

# 이분산 t-test (댓글수 비교) // 실험군, 대조군 (코드상)
t_stat_comments, p_value_comments = stats.ttest_ind(experimental_group_comment, control_group_comment, equal_var=False)

# 이분산 t-test (반응수 비교)
t_stat_reactions, p_value_reactions = stats.ttest_ind(experimental_group_reaction, control_group_reaction, equal_var=False)

# 결과 출력
print(f"댓글수에 대한 t-statistic: {t_stat_comments}, p-value: {p_value_comments}")
# 유의수준 0.05에서 결과 해석
if p_value_comments < 0.05:
    print(">>> 검증 결과 : 귀무가설을 기각합니다. 팬들은 실험군에 대해 더 많은 댓글을 남깁니다.")
else:
    print(">>> 검증 결과 : 귀무가설을 기각할 수 없습니다. 댓글수에 유의미한 차이가 없습니다.")

print(' --- 대조군 평균:', control_group_comment.mean(), '실험군 평균:', experimental_group_comment.mean())
print('')


print(f"반응수에 대한 t-statistic: {t_stat_reactions}, p-value: {p_value_reactions}")
if p_value_reactions < 0.05:
    print(">>> 검증 결과 : 귀무가설을 기각합니다. 팬들은 실험군에 대해 더 많은 반응을 보입니다.")
else:
    print(">>> 검증 결과 : 귀무가설을 기각할 수 없습니다. 반응수에 유의미한 차이가 없습니다.")

print(' --- 대조군 평균:', control_group_reaction.mean(), '실험군 평균:', experimental_group_reaction.mean())

print('')

print('=-----')
print('>>> 따라서,    --- 기각!!')
print('이 기각이 의미하는 바?')
```

- 대립가설: ~~ 차이가 없다 (대립가설 ) 세워서 해야하는거자나?

```python
df['게시글 내용_질문 여부'].value_counts()
```

```python
import pandas as pd
from scipy import stats

# 대조군  / 실험군 : 질문한거  (우리가 알고자 하는거)
 

experimental_group_comment = df[df['게시글 내용_질문 여부'] == 1]['댓글수'] # 실험군
control_group_comment = df[df['게시글 내용_질문 여부'] == 0]['댓글수'] # 대조군
print(len(experimental_group_comment), len(control_group_comment))
print('')

experimental_group_reaction = df[df['게시글 내용_질문 여부'] == 1]['반응수'] # 실험군
control_group_reaction = df[df['게시글 내용_질문 여부'] == 0]['반응수'] # 대조군
print(len(experimental_group_reaction), len(control_group_reaction))
print('')

# 이분산 t-test (댓글수 비교) // 실험군, 대조군 (코드상)
t_stat_comments, p_value_comments = stats.ttest_ind(experimental_group_comment, control_group_comment, equal_var=False)

# 이분산 t-test (반응수 비교)
t_stat_reactions, p_value_reactions = stats.ttest_ind(experimental_group_reaction, control_group_reaction, equal_var=False)

# 결과 출력
print(f"댓글수에 대한 t-statistic: {t_stat_comments}, p-value: {p_value_comments}")
# 유의수준 0.05에서 결과 해석
if p_value_comments < 0.05:
    print(">>> 검증 결과 : 귀무가설을 기각합니다. 댓글수에 유의미한 차이가 있습니다.")
else:
    print(">>> 검증 결과 : 귀무가설을 기각할 수 없습니다. 댓글수에 유의미한 차이가 없습니다.")

print(' --- 대조군 평균:', control_group_comment.mean(), '실험군 평균:', experimental_group_comment.mean())
print('')


print(f"반응수에 대한 t-statistic: {t_stat_reactions}, p-value: {p_value_reactions}")
if p_value_reactions < 0.05:
    print(">>> 검증 결과 : 귀무가설을 기각합니다. 반응수에 유의미한 차이가 있습니다.")
    print('')
    print('0.1까지 봐줄까')
else:
    print(">>> 검증 결과 : 귀무가설을 기각할 수 없습니다. 반응수에 유의미한 차이가 없습니다.")

print(' --- 대조군 평균:', control_group_reaction.mean(), '실험군 평균:', experimental_group_reaction.mean())

print('')

print('=-----')
print('차이는 없는데 ㅋㅋㅋ 오히려 질문한게 반응이 별로 없네 근데 좀 애매하긴 하다 이 분석이 맞나 ㅋㅋㅋ')

 
```
